```python
import tensorflow as tf
from tensorflow.keras.models import Model, load_model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
origin_model = tf.keras.models.load_model('/content/drive/MyDrive/Colab_Notebooks/model_2nd_version_0916.h5')

converter = tf.lite.TFLiteConverter.from_keras_model(origin_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
tflite_model = converter.convert()
```

```python
with open('/content/drive/MyDrive/Colab_Notebooks/tflite_model.tflite', 'wb') as f:
  f.write(tflite_model)
```
